# Linear Regression

Linear regression is a statistical method for modeling the relationship between a dependent variable and one or more independent variables. The goal is to find the best-fitting straight line that represents this relationship. The most common method to do this is Ordinary Least Squares (OLS), which minimizes the sum of the squared distances between the observed data points and the line.

The R 
2
  (R-squared) value is a key metric used to evaluate how well this linear model fits the data. It measures the proportion of the variance in the dependent variable that is predictable from the independent variable(s). An R 
2
  of 0.85, for example, means that 85% of the variability in the dependent variable can be explained by the model's independent variables.

## Importing Libraries

```python
try:
    import matplotlib.pyplot as plt
    import pandas as pd
    import numpy as np
    from scipy import stats
    
    print('Importing successful')

except ImportError as e:
    print(f"Import failed: {e}")
```

## Import Sample Data

```python
df = pd.read_csv('cars_data.csv')
print('Data importet succesfully')
```

```python
df.head(3)
```

```python
df.dtypes
```

## Extract dependent and independent variable and clean from missing values

```python
df_clean = df.dropna(subset=['year', 'price', 'engine_power'])
```

```python
x1 = df_clean['year']
y1 = df_clean['price']
```

```python
x1.head(3)
```

```python
y1.head(3)
```

## Scatter Plot of the variables to see if there is a linear Relationship

```python
plt.figure(figsize=(10, 6)) # Set the figure size for better visualization
plt.scatter(x1, y1, alpha=0.7) # Plot with some transparency
plt.title('Car Price vs. Year') # Title of the plot
plt.xlabel('Year') # Label for the x-axis
plt.ylabel('Price') # Label for the y-axis
plt.grid(True) # Add a grid for easier readability
plt.show()
```

## Calculate r_value, slope, intercept, p_value and std_err

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x1, y1)
```

```python
print("\nLinear Regression Results:")
print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"R-squared: {r_value**2:.2f}")
print(f"p-value: {p_value:.2f}")
print(f"Standard Error: {std_err:.2f}")
```

These results tell you about the relationship between a car's year and its price based on your data. Here is a breakdown of what each value means:

Slope: 441.49

This is the most important number. It means that for every one-year increase in a car's age, its price is predicted to increase by 441.49. This is the key takeaway: Cars that are a year newer are, on average, worth about $441 more.

Intercept: -879747.82

This is the predicted price of a car in the year zero. In this specific case, the number itself isn't meaningful because you can't have a car from the year zero. Its purpose is simply to define the starting point of the regression line on the y-axis.

R-squared: 0.18

This value indicates the strength of the relationship. An R-squared of 0.18 means that 18% of the variation in car prices can be explained by the car's year. This is a low value, which tells you that the year alone is not a very strong predictor of a car's price. Other factors like mileage, engine size, or condition are likely much more important.

p-value: 0.00

This value shows if the relationship is statistically significant. A p-value of 0.00 is very low (it's less than 0.05, the typical threshold). This means that the relationship between a car's year and its price is statistically significant, and the trend you're seeing is not due to random chance.

Standard Error: 15.69

This is a measure of the accuracy of your slope estimate. A smaller number here means that your calculated slope is a more reliable estimate of the true relationship between the variables.

## Show the Plot again with Regression Line

```python
# First, create an array of x-values that span the range of your data
x_predict = np.array([x1.min(), x1.max()])
# Then, calculate the corresponding y-values using the linear regression formula (y = slope*x + intercept)
y_predict = slope * x_predict + intercept
```

```python
plt.figure(figsize=(10, 6)) # Set the figure size for better visualization
plt.scatter(df['year'], df['price'], alpha=0.7, label='Actual Data Points') # Plot with some transparency
plt.title('Car Price vs. Year with Regression Line') # Title of the plot
plt.xlabel('year') # Label for the x-axis
plt.ylabel('Price') # Label for the y-axis
plt.grid(True) # Add a grid for easier readability

# Finally, plot the line on top of the scatter plot
plt.plot(x_predict, y_predict, color='red', linestyle='--', linewidth=2, label='Regression Line')

plt.legend() # Add a legend to distinguish the scatter points from the line
plt.show()
```

## Second example

```python
x2 = df_clean['engine_power']
y2 = df_clean['price']
```

```python
plt.figure(figsize=(10, 6)) # Set the figure size for better visualization
plt.scatter(x2, y2, alpha=0.7) # Plot with some transparency
plt.title('Car Price vs. Engine Power') # Title of the plot
plt.xlabel('Engine_Power') # Label for the x-axis
plt.ylabel('Price') # Label for the y-axis
plt.grid(True) # Add a grid for easier readability
plt.show()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(x2, y2)
```

```python
print("\nLinear Regression Results:")
print(f"Slope: {slope:.2f}")
print(f"Intercept: {intercept:.2f}")
print(f"R-squared: {r_value**2:.2f}")
print(f"p-value: {p_value:.2f}")
print(f"Standard Error: {std_err:.2f}")
```

Let's break down what each of these numbers tells you about the connection between a car's engine power and its price.

Slope: 5654.83

This is the most important result. It means that for every one-unit increase in engine_power, the car's price is predicted to increase by $5654.83, on average. This suggests a strong positive relationship: more powerful engines are associated with higher prices.

Intercept: 1376.80

The intercept is the predicted price when the engine power is zero. In this case, it doesn't have a practical meaning because a car can't have zero engine power. It just serves as the starting point for your regression line.

R-squared: 0.19

The R-squared value of 0.19 means that 19% of the variation in car prices can be explained by the car's engine power. This is slightly better than the R-squared for your previous model with year (which was 0.18), but it still shows that engine power alone isn't a very strong predictor. Many other factors are at play.

p-value: 0.00

Since your p-value is essentially zero (it's less than 0.05), the relationship between engine power and price is considered statistically significant. This means the trend you're seeing in your data is likely real and not just a random coincidence.

Standard Error: 191.87

This number tells you about the precision of your slope estimate. A smaller standard error means your slope of 5654.83 is a more reliable estimate.

These results suggest that engine_power is a slightly better predictor of price than year was, but there are still many other factors that influence a car's price.

```python
# First, create an array of x-values that span the range of your data
x_predict2 = np.array([x2.min(), x2.max()])
# Then, calculate the corresponding y-values using the linear regression formula (y = slope*x + intercept)
y_predict2 = slope * x_predict2 + intercept
```

```python
plt.figure(figsize=(10, 6)) # Set the figure size for better visualization
plt.scatter(df['engine_power'], df['price'], alpha=0.7, label='Actual Data Points') # Plot with some transparency
plt.title('Car Price vs. Engine Power with Regression Line') # Title of the plot
plt.xlabel('Engine_Power') # Label for the x-axis
plt.ylabel('Price') # Label for the y-axis
plt.grid(True) # Add a grid for easier readability

# Finally, plot the line on top of the scatter plot
plt.plot(x_predict2, y_predict2, color='red', linestyle='--', linewidth=2, label='Regression Line')

plt.legend() # Add a legend to distinguish the scatter points from the line
plt.show()
```
